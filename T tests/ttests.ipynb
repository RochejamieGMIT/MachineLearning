# Normal probability Funciton

$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{- \frac{1}{2} \big ( \frac {x - \mu}{\sigma} \big) ^2}  $


```python
import numpy as np

def normal_pdf (x, mu=0.0, sigma=1.0):
    #Answer A*B
    A = 1.0/(sigma *np.sqrt(2.0*np.pi))
    B = np.exp(-0.5*((x-mu)/sigma)**2)
    return A*B
```

```python
normal_pdf(0.0,mu=-0.0,sigma=np.sqrt(0.2))
```

## Plots

![normal PDF](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/330px-Normal_Distribution_PDF.svg.png)

```python
# blank plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
#
x = np.linspace(-5.0,5.0,1001)

# plot values

for mu,sigma2 in[[0.0,0.2],[0.0,1.0],[0.0,5.0],[-2.0,0.5]]:
    y = normal_pdf(x,mu=mu,sigma=np.sqrt(sigma2))
    ax.plot(x,y,label=f'$\mu {mu},\sigma^2 {sigma2}$')

ax.legend()
ax.grid()
```

# Area under the curve 



```python
fig, ax = plt.subplots(figsize=(12,6))
# PDF STD normal dist
x = np.linspace(-4.0,4.0,1001)

# plot values

for mu,sigma2 in[[0.0,1.0]]:
    y = normal_pdf(x,mu=mu,sigma=np.sqrt(sigma2))
    ax.plot(x,y,label=f'$\mu {mu},\sigma^2 {sigma2}$')
#add x axis
ax.axhline(0.0,color="black")

ax.vlines([-1.0,1],0.0,0.4,color="black")

x_interval = np.linspace(-1.0,1.0,101)
#shade the area under the blue curve
ax.fill_between(x_interval,normal_pdf(x_interval),color="green",alpha=0.4)

ax.legend()
```

```python
# the intersection of the gray and blue lines

print(f'(-1.0,{normal_pdf(-1.0)})')
print(f'(1.0,{normal_pdf(1.0)})')
```

# independent Sample T-Test

```python
# dataframes
import pandas as pd
import scipy.stats as ss

df = pd.read_csv("Iris.csv")

df
```

```python
# petal len into dataframe

sample_a = df[df["Species"]=="Iris-setosa"]["PetalLengthCm"].to_numpy()

sample_a
```

```python
sample_b = df[df["Species"]=="Iris-virginica"]["PetalLengthCm"].to_numpy()

sample_b
```

```python
# t-Test output

ss.ttest_ind(sample_a,sample_b)  
```
